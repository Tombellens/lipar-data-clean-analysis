```python
from dateutil import parser

def extract_year(date_str):
  date_str= translate_month(date_str)
  if date_str == 'Present':
    date_str = "2023"
  try:
    date_obj = parser.parse(date_str)
    return date_obj.year
  except (ValueError, OverflowError, TypeError):
    print('error parsing: ' + str(date_str))
    return None
```

```python
def translate_month(date_str):
  dutch_to_english_months  = {'januari': 'January', 'februari': 'February', 'maart': 'March', 'april': 'April', 'mei': 'May', 'juni': 'June', 'juli': 'July', 'augustus': 'August', 'september': 'September', 'oktober': 'October', 'november': 'November', 'december': 'December'}
  for dutch, english in dutch_to_english_months.items():
    try:
      if dutch in date_str:
        return date_str.replace(dutch, english)
    except:
        return date_str
  return date_str
```

```python
import ast

def convert_to_list(value):
    try:
        return ast.literal_eval(value)
    except (ValueError, SyntaxError):
        return [value]

positions_df['gov'] = positions_df['gov'].apply(convert_to_list)

# Function to merge rows with identical gov values
def merge_identical_gov(group):
    merged_rows = []
    for _, df in group.groupby(group['gov'].apply(str)):
        if df.shape[0] > 1:
            merged_row = df.iloc[0].copy()
            merged_row['start_year'] = df['start_year'].min()
            merged_row['end_year'] = df['end_year'].max()
            merged_rows.append(merged_row)
        else:
            merged_rows.append(df.iloc[0])
    return pd.DataFrame(merged_rows)

# Group by adviser_id and merge rows with identical gov values
grouped = positions_df.groupby('adviser_id').apply(merge_identical_gov).reset_index(drop=True)

# Explode rows with multiple gov values
exploded_rows = grouped[grouped['gov'].apply(len) > 1].explode('gov').reset_index(drop=True)

# Keep rows with single gov values
single_gov_rows = grouped[grouped['gov'].apply(len) == 1]

# Combine all rows into the new DataFrame
new_df = pd.concat([single_gov_rows, exploded_rows], ignore_index=True)

# Convert gov back to list for consistency
new_df['gov'] = new_df['gov'].apply(lambda x: [x] if isinstance(x, int) else x)

# Display the new DataFrame
new_df
```

```python
new_df['previous'] = None
new_df['next'] = None

# Function to find previous and next positions
def find_previous_next_positions(row, full_positions_df):
    adviser_positions = full_positions_df[(full_positions_df['adviser_id'] == row['adviser_id']) &
                                          (full_positions_df['position_id'] != row['pos_id'])]

    # Find previous positions
    previous_positions = adviser_positions[adviser_positions['end_year'] <= row['start_year']]
    previous_positions = previous_positions[previous_positions['end_year'] == previous_positions['end_year'].max()]

    # Find next positions
    next_positions = adviser_positions[adviser_positions['start_year'] >= row['end_year']]
    next_positions = next_positions[next_positions['start_year'] == next_positions['start_year'].min()]

    return previous_positions['position_id'].tolist(), next_positions['position_id'].tolist()

# Iterate over new_df and find previous and next positions
for idx, row in new_df.iterrows():
    prev, next_ = find_previous_next_positions(row, full_positions_df)
    new_df.at[idx, 'previous'] = prev
    new_df.at[idx, 'next'] = next_

# Display the new DataFrame
print(new_df)
```

```python
def find_labels(pos_ids):
    url = 'https://linkedin-parsing-49a53bd295b9.herokuapp.com/positions/'
    api_key = #API key
    headers = {
        'Authorization': f'{api_key}',
        'Content-Type': 'application/json',
    }

    labels = []
    for pos_id in pos_ids:
        # Making the GET request
        print(f"Fetching label for position ID: {pos_id}")
        response = requests.get(url + str(pos_id) + '/codes', headers=headers)

        # Checking the response
        if response.status_code == 200:
            print("Success:", response.json())
            label_found = False
            for code in response.json():
                if code['name'] == 'sector':
                    labels.append(code['value'])
                    label_found = True
                    break
            if not label_found:
                print('No sector code found for position ID:', pos_id)
                labels.append(None)
        else:
            print("Error fetching data for position ID:", pos_id)
            print("Error:", response.text)
            labels.append(None)

    return labels
```
